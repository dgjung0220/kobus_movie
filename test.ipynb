```python
import csv
#df['movieName']
column_names = ['movieName', 'director', 'company', 'distributor', 
              'openDt', 'type1', 'num_screen', 'num_sales', 'num_seats', 
              'seoul_sales', 'seoul_seats', 'genre', 'grade', 'type2']

from KobusMovie import KobusMovie

filename = './dataset.xlsx'
apikey = 'f5eef3421c602c6cb7ea224104795888'
kobusMovie = KobusMovie(filename, apikey)

refined_file = open('refined.csv', 'a', newline='', buffering=1)

wr = csv.writer(refined_file)

for idx, row in enumerate(kobusMovie.df.iterrows()):
    
    if idx<2464:
        continue
    
    data = row[1]
    movieName = data['movieName']
    directerNm = data['director']
    openDt = data['openDt']
       
    movieCd = kobusMovie.searchMovieCode(movieName, directerNm)
    
    if movieCd == 0:
        continue
    
    print(idx, movieName, movieCd)
    
    num_actors, num_staffs, running_time, top3_actors = kobusMovie.searchMovieInfo(movieCd)
    director_prev_movie_num = kobusMovie.count_movies_before_data(directerNm, openDt)
    
    if len(top3_actors) > 0:
        actors_average_seats_num = 0
        divide_num = len(top3_actors)
        
        for actor in top3_actors:        
            num = kobusMovie.get_people_counts_from_actor(actor, movieName, openDt)
            actors_average_seats_num += num
        actors_prev_seats_num = int(actors_average_seats_num / divide_num)
        
        wr.writerow([idx, data['movieName'], data['director'], data['company'], data['distributor'],
                    data['openDt'], data['type1'], data['type2'], data['genre'], data['grade'], 
                    num_actors, num_staffs, running_time, ','.join(top3_actors),
                    data['num_screen'], data['num_seats'], data['seoul_seats'], data['num_sales'], data['seoul_sales'], director_prev_movie_num, actors_prev_seats_num])

refined_file.close()
```
